```python
# load table from wikipedia
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_Pixar_films", header=0)

filmography_table = tables[0]
filmography_table.drop(axis=1, index=0, inplace=True)

# display table
filmography_table
```

```python
# drop irrelevant columns
filmography_table.drop(["Writer(s)", "Writer(s).1", "Producer(s)", "Composer(s)"], axis=1, inplace=True)
```

```python
# parse release date
filmography_table["Release date"] = filmography_table["Release date"].apply(lambda x: int(str(x)[-4:]))

filmography_table.iloc[0]
```

```python
# parse director column
filmography_table = filmography_table.rename(columns={"Director(s)": "Director"})

filmography_table["Director"] = filmography_table["Director"].apply(lambda x: x.split("Co-directed by")[0])

filmography_table.iloc[0]
```

```python
# manual fix for Brave (2012)
filmography_table.loc[filmography_table["Film"] == "Brave", "Director"] = "Brenda Chapman"
```

```python
# convert dataframe to exportable list
films = []

for film in filmography_table.values.tolist():
    films.append({
        "title": film[0],
        "year": film[1],
        "director": film[2]
    })

films[0]
```
