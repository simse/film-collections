```python
# load table from wikipedia
tables = pd.read_html("https://en.wikipedia.org/wiki/Steven_Spielberg_filmography", header=1)

filmography_table = tables[0]
```

```python
# display table
filmography_table
```

```python
# filter out table where he wasn't the director
filmography_table = filmography_table[filmography_table["Director"] == "Yes"] 
```

```python
# drop irrelevant columns
filmography_table = filmography_table.drop(["Director", "Writer", "Producer", "Notes", ".mw-parser-output .tooltip-dotted{border-bottom:1px dotted;cursor:help}Ref."], axis=1)
```

```python
# convert dataframe to exportable list
films = []

for film in filmography_table.values.tolist():
    films.append({
        "title": film[1],
        "year": film[0],
        "director": "Steven Spielberg"
    })
```
